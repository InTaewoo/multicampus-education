# 0812 목 : Data Cleansing

```python
# 1) 정제할 파일 불러오기

import pandas as pd
import numpy as np
df = pd.read_csv('apt.csv', encoding='utf-8')
```

```python
# 2) 거래 취소된 행 지우기

indexNames = df[ (df['no'] >= 20150101)
                & (df['no'] <= 202101212) ].index
df2 = df.drop(indexNames)
```

```python
df2
```

```python
# 3) 불필요한 컬럼 제거

df3 = df2.drop(columns=['add02','apt','floor','date','add03','when','no'])
```

```python
df3
```

```python
# 4) 중요한 컬럼 추출 (아파트 규모 분류)

def big_mid_small(x):
    if x>85:
        return 'big'
    elif x>60 and x<=85:
        return 'mid'
    else:
        return 'small'
    
    return 
df3['size2'] = df3['size'].apply(big_mid_small)
```

```python
df3
```

```python
# 5) 주소 ‘xx구’만 추출

def add01_add(x):
    return str(x)[6:9]
df3['add'] = df3['add'].apply(add01_add)
```

```python
df3
```

```python
# 6) 클렌징 된 데이터 저장

df3.to_csv('Apt_0812.csv', encoding='utf-8-sig')
df3.to_csv('Apt_kr_0812.csv', encoding='euc-kr')
```

### row 열 생성 후 값 부여하기

```python
def add_row(x):
    if x in ['강북구', '도봉구']:
        return '2'
    elif x in ['은평구','종로구','성북구','노원구']:
        return '3'
    elif x in ['서대문구','중구','동대문구','중랑구']:
        return '4'
    elif x in ['강구','마포구','용산구','성동구','광진구']:
        return '5'
    elif x in ['구로구','양천구','영등포구','동작구','강동구']:
        return '6'    
    else:
        return '7'
    
    return 
df3['row'] = df3['add'].apply(add_row)
```

### space 열 생성 후 값 부여하기


```python

def add_space(x):
    if x in ['강서구', '구로구']:
        return '4'
    elif x in ['은평구','서대문구','마포구','양천구','금천구']:
        return '5'
    elif x in ['강북구','종로구','중구','용산구','영등포구','관악구']:
        return '6'
    elif x in ['도봉구','성북구','동재문구','성동구','동작구','서초구']:
        return '7'
    elif x in ['노원구','중랑구','광진구','강동구','강남구']:
        return '8'    
    else:
        return '9'
    
    return 
df3['space'] = df3['add'].apply(add_space)
```
